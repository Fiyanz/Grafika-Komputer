```python
import matplotlib.pyplot as plt
```

```python
def brute_force(x1, y1, x2, y2):
    # Buat list untuk menyimpan koordinat hasil
    points = []
    
    # Hitung dx dan dy
    dx = x2 - x1
    dy = y2 - y1
    
    # Jika dx lebih besar dari dy, iterasi berdasarkan x
    if abs(dx) >= abs(dy):
        # Hitung kemiringan (slope) m
        m = dy / dx if dx != 0 else 0  # Menghindari pembagian dengan 0
        for x in range(x1, x2 + 1) if x1 < x2 else range(x1, x2 - 1, -1):
            # Hitung y menggunakan persamaan y = mx + c
            y = y1 + int(m * (x - x1))
            points.append((x, y))
    else:  # Jika dy lebih besar dari dx, iterasi berdasarkan y
        # Hitung kemiringan invers (1/m)
        m_inv = dx / dy if dy != 0 else 0  # Menghindari pembagian dengan 0
        for y in range(y1, y2 + 1) if y1 < y2 else range(y1, y2 - 1, -1):
            # Hitung x menggunakan persamaan x = (y - c) / m
            x = x1 + int(m_inv * (y - y1))
            points.append((x, y))

    return points
```

```python
def plot_brute_force(x1, y1, x2, y2):
    kordinat = brute_force(x1, y1, x2, y2)

    x_coords = [point[0] for point in kordinat]
    y_coords = [point[1] for point in kordinat]

    plt.figure(figsize=(6, 6))
    plt.plot(x_coords, y_coords, marker='o', color='blue', markersize=5, linestyle='-', linewidth=1)
    plt.title("Plot Titik Perpindahan Algoritma DDA")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
x1, y1, x2, y2 = 2, 1, 8, 5

grafik = plot_brute_force(x1, y1, x2, y2)
algo_bf = brute_force(x1, y1, x2, y2)
print(algo_bf)
print(grafik)
```
