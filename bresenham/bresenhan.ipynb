```python
import matplotlib.pyplot as plt
```

```python
def bresenham(x1, y1, x2, y2):
    # Hitung dx dan dy
    dx = x2 - x1
    dy = y2 - y1

    # Tentukan arah langkah
    x_step = 1 if dx > 0 else -1
    y_step = 1 if dy > 0 else -1
    dx = abs(dx)
    dy = abs(dy)

    # Buat list untuk menyimpan koordinat hasil
    points = []
    x, y = x1, y1

    # Tentukan apakah garis lebih horizontal atau vertikal
    if dx > dy:  # Lebih horizontal
        # Inisialisasi parameter keputusan
        p = 2 * dy - dx
        for _ in range(dx + 1):  # Loop sebanyak panjang dx
            points.append((x, y))  # Tambahkan titik saat ini ke list
            x += x_step
            if p >= 0:
                y += y_step
                p -= 2 * dx
            p += 2 * dy
    else:  # Lebih vertikal
        # Inisialisasi parameter keputusan
        p = 2 * dx - dy
        for _ in range(dy + 1):  # Loop sebanyak panjang dy
            points.append((x, y))  # Tambahkan titik saat ini ke list
            y += y_step
            if p >= 0:
                x += x_step
                p -= 2 * dy
            p += 2 * dx

    return points
```

```python
def plot_bresenhan(x1, y1, x2, y2):
    kordinat = bresenham(x1, y1, x2, y2)

    x_coords = [point[0] for point in kordinat]
    y_coords = [point[1] for point in kordinat]

    plt.figure(figsize=(6, 6))
    plt.plot(x_coords, y_coords, marker='o', color='blue', markersize=5, linestyle='-', linewidth=1)
    plt.title("Plot Titik Perpindahan Algoritma DDA")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
x1, y1, x2, y2 = 2, 1, 8, 5
algo_bres = bresenham(x1, y1, x2, y2)
graf_bres = plot_bresenhan(x1, y1, x2, y2)
print(algo_bres)
print(graf_bres)
```
